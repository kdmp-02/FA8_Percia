```python
#Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/kdmp-02/Example-Dataset/main/Salaries.csv")
```

```python
dfcount = df.groupby('rank')[['salary']].count()
dfmissing = df.groupby('rank')[['salary']].apply(lambda x: x.isna().sum())
dfmedian = df.groupby('rank')[['salary']].median()
dfmean = df.groupby('rank')[['salary']].mean()
dfstdmean = df.groupby('rank')[['salary']].sem()
dfstdv = df.groupby('rank')[['salary']].std()

groups = df.groupby('rank')
for group1 in groups:
    dfcv = dfstdv / dfmean
    
dfskew = df.groupby('rank')[['salary']].skew()
dfseskew = df.groupby('rank')[['salary']].apply(lambda x: ((6*len(x)*(len(x)-1))/((len(x)-2)*(len(x)+1)*(len(x)+3)))**(1/2))
dfkurtosis = df.groupby('rank')[['salary']].apply(lambda x:np.round(x.kurtosis(numeric_only=True), 3))


dfmin = df.groupby('rank')[['salary']].min()
dfmax = df.groupby('rank')[['salary']].max()
dfquantile25 = df.groupby('rank')[['salary']].quantile(0.25)
dfquantile50 = df.groupby('rank')[['salary']].quantile(0.5)
dfquantile75 = df.groupby('rank')[['salary']].quantile(0.75)
dfquantile10 = df.groupby('rank')[['salary']].quantile(0.1)
dfquantile20 = df.groupby('rank')[['salary']].quantile(0.2)
dfquantile30 = df.groupby('rank')[['salary']].quantile(0.3)
dfquantile40 = df.groupby('rank')[['salary']].quantile(0.4)
dfquantile50 = df.groupby('rank')[['salary']].quantile(0.5)
dfquantile60 = df.groupby('rank')[['salary']].quantile(0.6)
dfquantile70 = df.groupby('rank')[['salary']].quantile(0.7)
dfquantile80 = df.groupby('rank')[['salary']].quantile(0.8)
dfquantile90 = df.groupby('rank')[['salary']].quantile(0.9)


result = pd.concat([dfcount, dfmissing, dfmedian, dfmean, dfstdmean, dfstdv, dfcv, dfskew, dfseskew, dfkurtosis, dfmin, dfmax, dfquantile25, dfquantile50, dfquantile75, dfquantile10, dfquantile20, dfquantile30, dfquantile40, dfquantile50, dfquantile60, dfquantile70, dfquantile80, dfquantile90], axis=1).round(decimals=3)
result.columns = ['Valid', 'Missing', 'Median', 'Mean', 'Std. Error of Mean', 'Std. Deviation', 'Coefficient of variation', 'Skewness', 'Std. Error of Skewness', 'Kurtosis', 'Minimum', 'Maximum', '25th pecentile', '50th percentile', '75th percentile', '10th percentile', '20th percentile', '30th percentile', '40th percentile', '50th percentile', '60th percentile', '70th percentile', '80th percentile', '90th percentile']
result.transpose()
```
